```python
test_pose = (1.0, 0.0, 1.3)
print '====== PRUEBA ', 1, '======='
print 'Desired pose: \n', test_pose    
print '\nInverse Kinematics: '
ik_sol = grk.ik_position(*test_pose)
if ik_sol:
    print ik_sol

print '\nForward Kinematics: '
if ik_sol:
    print grk.forward_kinematics(*ik_sol[0])
else:
    print 'No solution found'
print '\n'
```

## Import poses from file

```python
# Read file and storage poses in 'test_pose' variable as tuples
f = open('ik_test_poses/left_arm_test_poses(1000).csv', 'r')
test_poses = []
for line in f:
    pose = []
    for string in line.split(','):
        pose.append(round(float(string),6))
    pose = tuple(pose)
    test_poses.append(pose)
test_poses = tuple(test_poses)
print len(test_poses), 'poses read!'
```

```python
for j in range(len(test_poses)):
    print '====== PRUEBA ', j+1, '======='
    print 'Desired pose: \n', test_poses[j]
    
    print '\nInverse Kinematics: '
    ik_sol = glk.inverse_kinematics(*test_poses[j])
    
    print '\nForward Kinematics: '
    if ik_sol:
        print glk.forward_kinematics(*ik_sol[0])
    else:
        print 'No solution found'
    print '\n'
```
